```python
import numpy as np
from skimage import io
import glob
```

```python
files = glob.glob("../input/train/*.jpg")
```

```python
train_imgs = []
for fname in files:
    img = io.imread(fname)
    train_imgs.append(img)
```

```python
train_imgs = np.array(train_imgs)
```

```python
print(type(train_imgs), train_imgs.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(10):
    plt.subplot(2,5, i+1)
    plt.imshow(train_imgs[i])
```

```python
tmp_label = []
for fname in files:
    tmp_label.append(fname.split('/')[3])
```

```python
labels = []
for fname in tmp_label:
    labels.append(fname.split('.')[0])
```

```python
import pandas as pd
```

```python
labels_dummy = pd.get_dummies(labels)
```
