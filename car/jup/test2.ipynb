```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../data/train.csv', index_col=0, na_values="?")
test = pd.read_csv('../data/test.csv', index_col=0, na_values="?")
```

```python
df.head()
```

```python
df.columns
```

```python
train = df[['cylinders', 'displacement', 'weight','acceleration', 'model year', 'origin']]
```

```python
Y = df[['mpg']]
```

```python
origin = train.pop('origin')
```

```python
train['USA'] = (origin == 1) *1.0
train['Europe'] = (origin == 2) * 1.0
train['Japan'] = (origin == 3) * 1.0
train.tail()
```

```python
trainX = train
```

```python
trainX.info()
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
```

```python
clf.fit(trainX, Y)
```

```python
test2X = test[['cylinders', 'displacement',  'weight', 'acceleration','model year', 'origin']]
```

```python
test2X
```

```python
origin = test2X.pop('origin')
```

```python
test2X['USA'] = (origin == 1) *1.0
test2X['Europe'] = (origin == 2) * 1.0
test2X['Japan'] = (origin == 3) * 1.0
test2X.tail()
```

```python
pred = clf.predict(test2X)
```

```python
sample = pd.read_csv('../data/sample_submit.csv',header=None)
```

```python
sample.head()
```

```python
sample[1] = pred
```
